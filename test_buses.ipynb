```python
import pandas as pd
from gtfs_functions import Feed

from DataStructures.Bus import BusTrip

# trips = pd.read_csv("data/CARTA_GTFS2024/trips.txt")
gtfs_path = "data/CARTA_GTFS.zip"

feed = Feed(gtfs_path)
trips = feed.trips
stops = feed.stops
stop_times = feed.stop_times
```

```python
stop_times
```

```python
trip_stops = stop_times.sort_values(by=['trip_id', 'stop_sequence']).groupby('trip_id')['stop_id'].apply(list).to_dict()
```

```python
trips_list = trips['trip_id'].to_list()
capacity = 50
bus_trips = []

for trip_id in trips_list:
    bus_trip = BusTrip(trip_id, trip_stops[trip_id], capacity)
    bus_trips.append(bus_trip)
```
